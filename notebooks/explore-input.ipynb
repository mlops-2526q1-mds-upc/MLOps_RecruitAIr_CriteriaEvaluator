```python
import torch
import pathlib
```

```python
for file in sorted(list(pathlib.Path("../data/raw/input-data").glob("*.pth"))):
    try:
        data = torch.load(str(file))
        print(f"Input Data: {file.name}")
        for k, v in data.items():
            if isinstance(v, torch.Tensor):
                print(f"  {k}: shape={v.shape} dtype={v.dtype} device={v.device}")
            elif isinstance(v, (list, tuple)):
                print(f"  {k}: {type(v).__name__.capitalize()} of length {len(v)}")
                for i, item in enumerate(v):
                    if isinstance(item, torch.Tensor):
                        print(f"    [{i}]: shape={item.shape} dtype={item.dtype} device={item.device}")
                    elif isinstance(item, str):
                        print(f"    [{i}]: str of length {len(item)}")
                    else:
                        print(f"    [{i}]: {type(item).__name__}")
            else:
                print(f"  {k}: {v}")
    except Exception as e:
        print(f"Could not load input data {file.name}: {e}")
```
